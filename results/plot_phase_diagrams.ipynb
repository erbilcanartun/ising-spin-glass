```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format='retina'
```

# bd
Renormalization procedure applying bond-moving first and then decimation

### Load and extract the data

```python
loaded_data = np.load('data/bd/boundary_data.npz')

hboundary_bd = loaded_data['hboundary']
vboundary_bd = loaded_data['vboundary']
```

### Plot the phase diagram

```python
plt.style.use('classic')
fig = plt.figure(figsize=(8, 6), dpi=100)
fig.set_facecolor('none')

lw = 2.5
fs = 20
plt.rc('lines', linewidth=lw)
plt.rc('axes', linewidth=lw)

ax = plt.subplot(111)
ax.plot(hboundary_bd[:,0], hboundary_bd[:,2], ls='-', lw=lw, marker='', ms=10, c='k')
ax.plot(hboundary_bd[:,1], hboundary_bd[:,2], ls='-', lw=lw, marker='', ms=10, c='k')
ax.plot(vboundary_bd[:,0], vboundary_bd[:,2], ls='-', lw=lw, marker='', ms=10, c='k')
ax.plot(vboundary_bd[:,1], vboundary_bd[:,2], ls='-', lw=lw, marker='', ms=10, c='k')

ax.axis([0, 1, 0, 30])
ax.tick_params(axis="both", direction="in", left=True,
               width=lw, length=4, labelsize=fs)
ax.set_xlabel("Antiferromagnetic Bond Concentration p", fontsize=fs)
ax.set_ylabel(r"Temperature $\mathrm{1/J}$", fontsize=fs)

#ax.text(0.48, 10, "D", fontsize=fs)
#ax.text(0.07, 4, "F", fontsize=fs)
#ax.text(0.9, 4, "AF", fontsize=fs)
#ax.text(0.47, 0.2, "SG", fontsize=fs)

plt.tight_layout()
#plt.savefig("phase_diagram_bd", dpi=300)
plt.show()
```

# db
Renormalization procedure applying decimation first and then bond-moving

### Load and extract the data

```python
loaded_data = np.load('data/db/boundary_data.npz')

hboundary_db = loaded_data['hboundary']
vboundary_db = loaded_data['vboundary']
```

### Plot the phase diagram

```python
plt.style.use('classic')
fig = plt.figure(figsize=(8, 6), dpi=100)
fig.set_facecolor('none')

lw = 2.5
fs = 20
plt.rc('lines', linewidth=lw)
plt.rc('axes', linewidth=lw)

ax = plt.subplot(111)
ax.plot(hboundary_db[:,0], hboundary_db[:,2], ls='-', lw=lw, marker='', ms=10, c='k')
ax.plot(hboundary_db[:,1], hboundary_db[:,2], ls='-', lw=lw, marker='', ms=10, c='k')
ax.plot(vboundary_db[:,0], vboundary_db[:,2], ls='-', lw=lw, marker='', ms=10, c='k')
ax.plot(vboundary_db[:,1], vboundary_db[:,2], ls='-', lw=lw, marker='', ms=10, c='k')

ax.axis([0, 1, 0, 3])
ax.tick_params(axis="both", direction="in", left=True,
               width=lw, length=4, labelsize=fs)
ax.set_xlabel("Antiferromagnetic Bond Concentration p", fontsize=fs)
ax.set_ylabel(r"Temperature $\mathrm{1/J}$", fontsize=fs)

ax.text(0.48, 2, "D", fontsize=fs)
ax.text(0.07, 1, "F", fontsize=fs)
ax.text(0.9, 1, "AF", fontsize=fs)
ax.text(0.47, 0.4, "SG", fontsize=fs)

plt.tight_layout()
#plt.savefig("phase_diagram_db", dpi=300)
plt.show()
```
